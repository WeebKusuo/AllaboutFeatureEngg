Nominal Features & One Hot Encoding — Notes
1. Nominal Feature (Categorical Feature)

Nominal feature wo hota hai jisme distinct categories hote hain (no order).
Example: color = {red, green, blue}

2. One Hot Encoding (OHE)

Categorical values ko numerical binary features me convert karta hai.

Har category ke liye ek naya feature/column ban jata hai.

Example:

Feature: color = red, green, blue

One Hot Encoding banayega:

red	green	blue
1	0	0
0	1	0

Agar color = red, to red = 1, aur baaki sab = 0.

3. One Hot Encoding — Disadvantages
❌ High Dimensionality

Agar feature me 100 categories hain → OHE 100 new columns banayega.

Data ka size aur dimensionality bahut badh jati hai.

❌ Sparse Matrix Problem

Zyada tar values 0 hoti hain → matrix sparse ban jaati hai.

Memory aur computation cost badh jaata hai.

❌ Overfitting ka Risk

Model training data ko unnecessary detail tak fit kar leta hai.

Generalization (new data par performance) kharab ho sakti hai.

4. Conclusion

OHE simple aur effective hai jab categories kam ho.

Large number of categories wale features ke liye One Hot Encoding avoid karna chahiye.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# creating as simple datset

df=pd.DataFrame({
    'color':['red','blue','green','red','blue']
})
```

```python
df.head()
```

```python
##Create an instance of ONehotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
### perform fit and transform 
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
##for any new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```
